```python
import os
from pydub import AudioSegment

# Define the input and output directories
input_directory = "/content/drive/MyDrive/New_music"
output_directory = "/content/drive/MyDrive/New_trimmed"

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Function to split audio files into multiple 10-second segments
def split_audio(input_path, output_dir, segment_duration=10):
    try:
        # Load the audio file
        audio = AudioSegment.from_file(input_path, format="mp3")
        # Calculate the number of segments
        total_duration = len(audio)  # Total duration in milliseconds
        segment_count = total_duration // (segment_duration * 1000)

        for i in range(segment_count):
            # Extract the segment
            start_time = i * segment_duration * 1000
            end_time = start_time + segment_duration * 1000
            segment = audio[start_time:end_time]

            # Generate output file name
            base_name = os.path.splitext(os.path.basename(input_path))[0]
            output_file = os.path.join(output_dir, f"{base_name}_part{i + 1}.mp3")

            # Export the segment
            segment.export(output_file, format="mp3")
            print(f"Saved segment: {output_file}")
    except Exception as e:
        print(f"Error processing {input_path}: {e}")

# Traverse the dataset folder
for class_folder in os.listdir(input_directory):
    class_path = os.path.join(input_directory, class_folder)
    if os.path.isdir(class_path):  # Check if it's a folder
        # Create a corresponding folder in the output directory
        output_class_path = os.path.join(output_directory, class_folder)
        os.makedirs(output_class_path, exist_ok=True)

        # Process each file in the class folder
        for file_name in os.listdir(class_path):
            file_path = os.path.join(class_path, file_name)
            if os.path.isfile(file_path) and file_name.endswith('.mp3'):  # Process only mp3 files
                # Split and save the audio segments
                split_audio(file_path, output_class_path)
```

```python
from pydub import AudioSegment
import os

# Path to the input mp3 file and output directory
input_file = "/content/MAYA HASEKO - Basanta Thapa  Shanti Shree  Pralhad Subedi Nepali Song 2077 Durgesh Thapa  Losina.mp3"
output_directory = "/content/drive/MyDrive"

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Function to split audio file into 5-second segments
def split_audio(input_path, output_dir, segment_duration=5):
    try:
        audio = AudioSegment.from_file(input_path, format="mp3")
        total_duration = len(audio)
        segment_count = total_duration // (segment_duration * 1000)

        for i in range(segment_count):
            start_time = i * segment_duration * 1000
            end_time = start_time + segment_duration * 1000
            segment = audio[start_time:end_time]

            base_name = os.path.splitext(os.path.basename(input_path))[0]
            output_file = os.path.join(output_dir, f"{base_name}_part{i + 1}.mp3")
            segment.export(output_file, format="mp3")
            print(f"Saved segment: {output_file}")

    except Exception as e:
        print(f"Error processing {input_path}: {e}")

# Process the single file
split_audio(input_file, output_directory)
```

To count files

```python
import os

# Path to the directory containing genre folders
data_dir = '/content/drive/MyDrive/trimmed'  # Update with your directory path

# Dictionary to store counts
mp3_counts = {}

# Loop through each subfolder (genre)
for genre in os.listdir(data_dir):
    genre_path = os.path.join(data_dir, genre)

    if os.path.isdir(genre_path):  # Check if it's a directory
        # Count .mp3 files in the subfolder
        mp3_files = [file for file in os.listdir(genre_path) if file.endswith('.mp3')]
        mp3_counts[genre] = len(mp3_files)

# Print the results
for genre, count in mp3_counts.items():
    print(f"{genre}: {count} mp3 files")
```

converting directory to single flat structure along with index name

```python
import os
import shutil

# Directory structure
source_dir = '/content/drive/MyDrive/trimmed'  # Original directory where your genre-based subfolders are
target_dir = '/content/drive/MyDrive/New_indexes_mp3'  # Directory where you want to create the flat structure

# Create target directory if it doesn't exist
if not os.path.exists(target_dir):
    os.makedirs(target_dir)

# Initialize lists for storing file paths and genre labels
audio_files = []
labels = []

# Iterate over genre-based subdirectories
for genre in os.listdir(source_dir):
    genre_dir = os.path.join(source_dir, genre)

    # Only process directories (not files)
    if os.path.isdir(genre_dir):
        # Iterate over each file in the genre directory
        for idx, filename in enumerate(os.listdir(genre_dir)):
            if filename.endswith('.mp3') or filename.endswith('.wav'):  # Include mp3 and wav files
                file_path = os.path.join(genre_dir, filename)

                # Create a new filename based on index and genre (e.g., 'rock_1.mp3')
                new_filename = f"{genre}_{idx+1}.mp3"  # You can modify the extension as needed

                # Define the new path in the target flat structure
                new_file_path = os.path.join(target_dir, new_filename)

                # Move the file to the target directory with the new name
                shutil.copy(file_path, new_file_path)  # You can use shutil.move() to move instead of copying

                # Append the file path and label to lists
                audio_files.append(new_file_path)
                labels.append(genre)

print(f"Files have been organized into {target_dir}.")


```
